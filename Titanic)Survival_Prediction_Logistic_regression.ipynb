Import Libraries

```python
import pandas as pd
import numpy as np
from  sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
```

Load Datset

```python
df= pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/Titanic-Dataset.csv')
```

```python
df.head()
```

Data **Preprocessing**

```python
df.describe()
```

```python
df.shape
```

```python
df.columns
```

```python
df.drop(['PassengerId','SibSp','Embarked','Ticket','Cabin', 'Name', 'Parch', 'Fare',],axis=1,inplace=True)
```

```python
df.head()
```

```python
#Change Column name 'Sex' to 'Gender'
df.rename(columns={'Sex': 'Gender'}, inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
#Check missing value
df['Age'].isnull().sum()
```

```python
df['Age']= df['Age'].fillna(df['Age'].mean())
```

```python
df['Gender'].isnull().sum()
```

```python
df['Pclass'].isnull().sum()
```

```python
df['Survived'].isnull().sum()
```

Feature Scaling

```python
df['Gender'].unique()
```

```python
#Apply Label  Encoding

le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
```

```python
df.head()
```

```python
#Separate Independent and Dependent Features
X = df.drop('Survived', axis=1)
y = df['Survived']
```

```python
# Apply Standard Scalling
from sklearn.preprocessing import StandardScaler
Scalar = StandardScaler()
X_scaled = Scalar.fit_transform(X)
```

Split Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

Train Model

```python

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Impute missing 'Age' values with the mean
X_train = np.nan_to_num(X_train, nan=np.nanmean(X_train))
X_test = np.nan_to_num(X_test, nan=np.nanmean(X_test))
```

Make Prediction

```python

#prediction on Test Data
y_pred = model.predict(X_test)
print(y_pred)
```

Evaluate Model

```python

accuracy  = accuracy_score(y_test, y_pred)
Cmatrix = confusion_matrix(y_test, y_pred)
Creport = classification_report(y_test, y_pred)

print(accuracy)
print(Cmatrix)
print(Creport)
```

Test On new Data

```python
# Test On new Data
new_passengers = pd.DataFrame({
    'Pclass': [1, 3, 2],
    'Gender': [0, 1, 0],  # 0 for female, 1 for male
    'Age': [30, 25, 40]
})

# Apply the same scaling as the training data
new_passengers_scaled = Scalar.transform(new_passengers)

# Make predictions
new_predictions = model.predict(new_passengers_scaled)

# Print the predictions
print(new_predictions)
```

Checking accuracy with KNN Model

```python
#KNN

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
knn_pred = knn.predict(X_test)
accuracy  = accuracy_score(y_test, knn_pred)
print(accuracy)
```
